```python
import hazelcast

client = hazelcast.HazelcastClient(cluster_name="lab2")
topic = client.get_topic("my-distributed-topic").blocking()
map = client.get_map("RequestsCountMap").blocking() 
queue = client.get_queue("queue").blocking() 
```

```python
for key in range(0, 1000): map.put(key, key)
```

```python
from time import sleep

for i in range(100): 
    topic.publish("Message " + str(i))
    sleep(0.1)
client.shutdown()
```

```python
for i in range(100):
    queue.put(i)
    print("Producing " + str(i))
queue.put(-1)
print("Producer finished")
client.shutdown()
```
